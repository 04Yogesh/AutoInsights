```python
# Boxplot
for column in df.columns:
    if df[column].dtype == 'int64' or df[column].dtype == 'float64':
        plt.figure(figsize=(25,6))
        sns.boxplot(df[column], color='Salmon')
        plt.show()   
```

```python
for column in df.columns:
    if df[column].dtype == 'int64' or df[column].dtype == 'float64':
        plt.figure(figsize=(20,8))
        sns.distplot(df[column])
        plt.show()
```
